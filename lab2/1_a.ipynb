## Q11

```python
ch = input("Enter alphabet: ")

if 'a' <= ch <= 'z':
    print(chr(ord(ch) - 32))
elif 'A' <= ch <= 'Z':
    print(chr(ord(ch) + 32))
else:
    print("Not a letter")

# i/p = M
```

## Q12

```python
n = int(input("Enter number of terms: "))

a, b = 0, 1

for i in range(n):
    print(a, end=" ")
    a, b = b, a + b

# i/p = 3
```

## Q13

```python
n = int(input("Enter number: "))
i = 2
factors = set()

while n > 1:
    if n % i == 0:
        factors.add(i)
        n //= i
    else:
        i += 1

print("Prime factors:", *factors)

# i/p = 24
```

## Q14

```python
n = int(input("Enter decimal number: "))
choice = int(input("1-Binary 2-Octal 3-Hexadecimal: "))

match choice:
    case 1:
        print(bin(n)[2:])
    case 2:
        print(oct(n)[2:])
    case 3:
        print(hex(n)[2:].upper())
    case _:
        print("Invalid option")

# i/p = 3 to binary
```

## Q15

```python
import random

number = random.randint(1, 50)

while True:
    guess = int(input("Guess number (1–50): "))

    if guess == number:
        print("Correct! You win!")
        break
    elif guess < number:
        print("Too low!")
    else:
        print("Too high!")

# i/p = 45, 46, 48
```

## Q16

```python
rows = int(input("Enter number of rows: "))

for i in range(1, rows + 1):
    for j in range(rows - i):
        print(" ", end="")

    for k in range(2*i - 1):
        print("*", end="")
    
    print() 


# i/p = 5
```
